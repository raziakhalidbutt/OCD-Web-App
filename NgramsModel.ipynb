```python
import nltk
from nltk.stem import WordNetLemmatizer
def lem(text):
  wordnet_lemmatizer = WordNetLemmatizer()
  return " ".join([wordnet_lemmatizer.lemmatize(word,pos="v") for word in text.split()])
```

```python
import spacy
sp = spacy.load('en_core_web_sm')

# sp.Defaults.stop_words -= {'get','other','being','how','against','someone','not','several','many','nothing','unless','before','same','one','no','do','enough','front','someone'}

filename = '/content/drive/MyDrive/Upwork Work/OCD/stopwordaddlist.txt'
add_list =set(functions_obj.readFile(filename))
sp.Defaults.stop_words |= (add_list) 

all_stopwords = sp.Defaults.stop_words  

sw_list = {'get','other','being','how','against','not','several','many','nothing','unless','before','same','one','no','do','enough','front','someone'}
all_stopwords_spacy = all_stopwords.difference(sw_list)




def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in all_stopwords_spacy])
```

# **Genism Pharaser package to create ngrams**

```python
#to create phrases
from gensim.models.phrases import Phrases, Phraser

pbigram = Phrases(nostopwords_unigrams, min_count=2, threshold=3) # higher threshold fewer phrases.(min count is delta)
ptrigram = Phrases(pbigram[nostopwords_unigrams], min_count=2, threshold=3)  
pfourgram = Phrases(ptrigram[pbigram[nostopwords_unigrams]], min_count=2, threshold=3)  

bigram_mod = Phraser(pbigram)
trigram_mod = Phraser(ptrigram)
fourgram_mod = Phraser(pfourgram)


def make_bigrams(texts):
    return [bigram_mod[word] for word in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[word]] for word in texts]

def make_fourgrams(texts):
    return [fourgram_mod[trigram_mod[bigram_mod[word]]] for word in texts]

#created ngrams
phraser_bigrams = make_bigrams(nostopwords_unigrams)
phraser_trigrams = make_trigrams(nostopwords_unigrams)
phraser_fourgrams = make_fourgrams(nostopwords_unigrams)
```

```python
import re
matches = []
for i in phraser_fourgrams:
  # print(i)
  # r = re.compile(".*_")
  r = re.compile(r"^([^_]*_){4}")
  newlist = list(filter(r.match, i)) # Read Note below
  # print(newlist)
  if newlist != []:
    matches.append(newlist)

# print(matches)
```

```python
def words_in_corpus(grams,dictionary):
      vocabs = {}                #creating dictionary with keys as word  and values as count of word in corpus
      for i in grams:            #loop through each sentence 
        # for j in i:              #loop through each word in sentence
          if i not in vocabs:    #if word not in dictionary
            vocabs[i]  = 1       #add word and set count to 1
          else:                  #if word not found in vocab
            vocabs[i] += 1 

      iv = []                    #list for words found in corpus
      oov  =  []                 #list for words not found in corpus
      for tok in dictionary:     #loop through the list of dictionary words
        if tok in vocabs:        #if word found in corpus vocabulary
          # print(tok,vocabs[tok])
          iv.append(tok)         #add it to the iv
        else:                    #if not found in corpus vocabulary
          oov.append(tok)        #add it to oov
      return iv,oov,vocabs


```

```python
def words_in_model_vocab(grams,dictionary):
      """ 
    
      grams: model ngrams (unigrm/bigram/trigram/fourgram)
      dictionary: list of ngrams (unigrm/bigram/trigram/fourgram)

      retrun In vocabulary words , Out of vocabulary words and corpus vocabulary
      """ 
      
      vocabs = {}                #creating dictionary with keys as word  and values as count of word in corpus
      for i in grams:            #loop through each word
        
          if i not in vocabs:    #if word not in dictionary
            vocabs[i]  = 1       #add word and set count to 1
          else:                  #if word not found in vocab
            vocabs[i] += 1       #increment the count of that word

      iv = []                    #list for words found in corpus
      oov  =  []                 #list for words not found in corpus
      for tok in dictionary:     #loop through the list of dictionary words
        if tok in vocabs:        #if word found in corpus vocabulary
          # print(tok,vocabs[tok])
          iv.append(tok)         #add it to the iv
        else:                    #if not found in corpus vocabulary
          oov.append(tok)        #add it to oov
      return iv,oov,vocabs
```
